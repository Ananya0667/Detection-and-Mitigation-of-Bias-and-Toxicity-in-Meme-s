```python
from transformers import pipeline
model_path = "cardiffnlp/twitter-roberta-base-sentiment-latest"
sentiment_task = pipeline("sentiment-analysis", model=model_path, tokenizer=model_path, return_all_scores=True)
sentiment_task("Covid cases are increasing fast!")
```

```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

def show_image(image_path):
    """
    Display an image given its file path using matplotlib.

    Args:
        image_path (str): The file path to the image.
    """
    try:
        # Load the image file
        img = mpimg.imread(image_path)

        # Display the image using imshow
        plt.imshow(img)
        plt.axis('off')  # Turn off axes (optional)
        plt.show()

    except FileNotFoundError:
        print(f"Error: The file '{image_path}' was not found.")
    except Exception as e:
        print(f"An error occurred: {e}")
```

# Mitigating Bais & Toxicity

## Method 1

using a classifer to classify the response generated by the LLLM


if negative score > 0.33 and positive score < 0.1 -> (bias and toxicity) response.
else normal response

```python
def response_checker(llm_reponse):
  scores = sentiment_task(llm_reponse)
  scores = scores[0]
  if scores[0]["score"] > 0.33 and scores[2]["score"] < 0.1:
    llm_reponse = "Sorry but as an AI model, i can't have bias and toxicity."
    return llm_reponse
  else:
    return llm_reponse
```

## Method 2

using a classifer to classify the user prompt before sending it to LLM


if positive score < 0.5 and abs(negative score - nuetral score) <= 0.15 -> (bias and toxicity) response.

else normal response

```python
def prompt_checker(llm_reponse):
  scores = sentiment_task(llm_reponse)
  scores = scores[0]
  if scores[2]["score"] < 0.5 and abs(scores[0]["score"] - scores[1]["score"]) <= 0.15:
    prompt_type = "hateful"
    return prompt_type
  else:
    return "not hateful"
```

```python
image_path = '/content/jew_people_meme.png'
image = Image.open(image_path)
enc_image = model.encode_image(image)
print("Image:")
show_image(image_path)
print()
for index, prompt in enumerate(test_prompts):
  print(f"Prompt {index + 1}: {prompt}")
  if prompt_checker(prompt) == "hateful":
    print(f"Response (after mitigating bias and toxicity): Sorry but as an AI model, i can't have bias and toxicity.\n")
  else:
    response = model.answer_question(enc_image, prompt, tokenizer)
    print(f"Original Response: {response}")
```

## Method 3

using a hateful meme detection model (a filter that checks the image itself if it is contains any thing hateful or not)

if score > 0.6 then hateful meme

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

hateful_image = pipeline("image-classification", model="tommilyjones/vit-base-patch16-224-finetuned-hateful-meme-restructured")
```

```python
def hateful_image_detector(image):
  scores = hateful_image(image)
  if scores[0]["score"] > 0.6:
    return "hateful"
  else:
    return "not hateful"
```

```python
image_path = '/content/jew_people_meme.png'
image = Image.open(image_path)
enc_image = model.encode_image(image)
print("Image:")
show_image(image_path)
print()
image_status = hateful_image_detector(image)
print(f"Detected: {image_status}")
if image_status == "not hateful":
  for index, prompt in enumerate(test_prompts):
    print(f"Prompt {index + 1}: {prompt}")
    response = model.answer_question(enc_image, prompt, tokenizer)
    print(f"Original Response: {response}")
else:
  print(f"Response (after mitigating bias and toxicity): Sorry but as an AI model, i can't have bias and toxicity.\n")
```
